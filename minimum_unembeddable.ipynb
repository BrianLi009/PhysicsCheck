```python
def check_subgraph(bigger_graphs, smaller_graphs):
    """bigger_graphs and smaller_graphs are a list of g6_string"""
    print ("looking at " + str(len(bigger_graphs)) + " graphs")
    non_minimal = []
    minimal = []
    for big_graph in bigger_graphs:
        graph1 = Graph(big_graph)
        for small_graph in smaller_graphs:
            graph2 = Graph(small_graph)
            h1 = graph1.subgraph_search(graph2)
            if h1 != None:
                print (small_graph + " is a subgraph of " + big_graph)
                non_minimal.append(big_graph)
                #h1.show()
    print ("showing non_minimal")
    print (non_minimal)
    for graph in bigger_graphs:
        if graph not in non_minimal:
            minimal.append(graph)
    print ("showing minimal")
    print (minimal)
    print ("In total " + str(len(non_minimal)) + " graphs")
```
