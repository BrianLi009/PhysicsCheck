```python
from z3 import *
import itertools
```

```python
def output_clause(edge_lst, n):
    """given the list of edges of a graph, we output the system of equations for embedability checking"""
    variable_dict = {}
    edge_to_var = {}
    setup = ''
    clauses = ''
    v_lst = range(0, n) #we can either do (0, n) or (1, n+1) depending on whether the edge list starts from 0 or 1
    for v in v_lst: 
        x_v = 'x' + '_' + str(v)
        y_v = 'y' + '_' + str(v)
        z_v = 'z' + '_' + str(v)
        variable_dict[x_v] = x_v
        variable_dict[y_v] = y_v
        variable_dict[z_v] = z_v
        edge_to_var[v] = [x_v, y_v, z_v]
    variables = list(variable_dict.keys())
    for variable in variables:
        clause = variable + ' = Real(' + '\''  + variable_dict[variable] + '\'' + ')' + '\n'
        setup = setup + clause
    print (setup)
    print ('s = Solver()')
    for i in range (len(variables)):
        if i%3 == 0: #a new vector
            x_i = variables[i]
            y_i = variables[i+1]
            z_i = variables[i+2]
            equation_1 = "s.add(" + x_i + "**2+" + y_i + "**2+" + z_i + "**2" + " == 1" + ")" + '\n'
            equation_2 = "s.add(" + z_i + ' >= 0' + ")" + '\n'
            equation_3 = "s.add(" + "Implies(" + z_i + " == 0, " + y_i + " > 0))" + '\n'
            equation_4 = "s.add(" + "Implies(And(" + z_i+ " == 0, " + y_i + " == 0), " + x_i + " == 1))" + "\n"
            clauses = clauses + equation_1 + equation_2 +equation_3 +equation_4
    for edge in edge_lst:
        vertex_1 = edge_to_var[edge[0]]
        vertex_2 = edge_to_var[edge[1]]
        equation_5 = "s.add(" + vertex_1[0] +'*'+vertex_2[0]+'+'+ vertex_1[1] +'*'+vertex_2[1]+'+'+vertex_1[2]+'*'+vertex_2[2]+' == 0' + ")" + '\n'
        clauses = clauses + equation_5
    for pair in list(itertools.combinations(v_lst, 2)):
        v_1 = pair[0]
        v_2 = pair[1]
        x_1 = variables[v_1 * 3]
        y_1 = variables[v_1 * 3+1]
        z_1 = variables[v_1 * 3+2]
        x_2 = variables[v_2 * 3]
        y_2 = variables[v_2 * 3+1]
        z_2 = variables[v_2 * 3+2]
        equation_6 = "s.add(" + "Or(Not(" + x_1 + " == " + x_2 + "), Not(" + y_1 + " == " + y_2 + "), Not(" + z_1 + " == " + z_2 + "))) \n"
        clauses = clauses + equation_6
    print (clauses)
    #print ("print (s.check())"+"\n")
    #print ("print (s.model())"+"\n")
```
